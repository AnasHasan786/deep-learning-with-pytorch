### **2. Tensor Shapes**

```python
x = torch.tensor([[1, 2, 3], [4, 5, 6]])
print("Tensor x: ", x)
print("Shape of x: ", x.shape)
```

```python
# It is used when we have to create a tensor of the same shape as x shape
torch.empty_like(x)
```

```python
# It is used when we have to create a tensor of the same shape as x shape
torch.zeros_like(x)
```

```python
# It is used when we have to create a tensor of the same shape as x shape
torch.ones_like(x)
```

```python
# torch.rand_like(x)

# Above command won't work as the datatype of x elements is int and the rand or rand_like generate numbers of float datatype.

## Solution for above command

torch.rand_like(x, dtype=torch.float32)
```

### **4. Mathematical Operations**

```python
# Scalar operation
x = torch.rand(2, 2)
x
```

```python
# addition
add = x + 2

print(add)
```

```python
# subtration
sub = x - 2

print(sub)
```

```python
# multiplication
mul = x * 2

print(mul)
```

```python
# division
div = x / 2

print(div)
```

```python
# integer division
int_div = (x*100) // 2

print(int_div)
```

```python
# mod
mod = (x*100) % 2

print(mod)
```

```python
# power
power = x**2

print(power)
```

### **5. Element wise operations**

```python
a = torch.rand(2, 3)
b = torch.rand(2, 3)

print(a)
print(b)
```

```python
# add
add = a + b

print(add)
```

```python
# sub
sub = a - b

print(sub)
```

```python
# mul
mul = a * b

print(mul)
```

```python
# div
div = a / b

print(div)
```

```python
# power
power = a**b

print(power)
```

```python
# mod
mod = a % b

print(mod)
```

```python
c = torch.tensor([1, -2, 3, -4])
```

```python
# abs
abs = torch.abs(c)

print(abs)
```

```python
# negative
neg = torch.neg(c)

print(neg)
```

```python
d = torch.tensor([1.9, 2.3, 3.7, 4.4])
```

```python
# round
round = torch.round(d)

print(round)
```

```python
# ceil
ceil = torch.ceil(d)

print(ceil)
```

```python
# floor
floor = torch.floor(d)

print(floor)
```

```python
# clamp
clamp = torch.clamp(d, min=2, max=3)

print(clamp)
```

### **7. Matrix operations**

```python
f = torch.randint(size=(2, 3), low=0, high=10)
g = torch.randint(size=(3, 2), low=0, high=10)

print(f)
print(g)
```

```python
# matrix multiplication
torch.matmul(f, g)
```

```python
vector1 = torch.tensor([1, 2])
vector2 = torch.tensor([3, 4])

# dot product
torch.dot(vector1, vector2)
```

```python
# transpose
torch.transpose(f, 0, 1)
```

```python
h = torch.randint(size=(3, 3), low=0, high=10, dtype=torch.float32)
h
```

```python
# determinant
torch.det(h)
```

```python
# inverse
torch.inverse(h)
```

### **8. Comparison Operations**

```python
i = torch.randint(size=(2, 3), low=0, high=10)
j = torch.randint(size=(2, 3), low=0, high=10)

print(i)
print(j)
```

```python
# greater than
i > j
```

```python
# less than
i < j
```

```python
# equal to
i == j
```

```python
# not equal to
i != j
```

```python
# greater than equal to
i >= j
```

```python
# less than equal to
i <= j
```
